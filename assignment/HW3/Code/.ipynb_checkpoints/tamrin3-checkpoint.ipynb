```python
# importing library that i needed
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
register_matplotlib_converters()
import yfinance as yf
from datetime import datetime, timedelta ,timezone
from statsmodels.tsa.arima.model import ARIMA
from time import time
```

# part1

```python
# 5 coins that have higher market cap 
cryptos = ['BTC-USD', 'ETH-USD','USDT-USD','BNB-USD','XRP-USD']
# yahoo finance just support 1d and longer times so we do not have 4h and 1h period of time
# timeframes = ['1d', '4h', '1h']
timeframes = ['1d']

# Date range , i get it abit longer than 1 year
start_date = '2022-9-01'
end_date = '2023-12-01'
```

# ADF test

```python
def perform_adf_test(series):
    result = adfuller(series)
#     print('ADF Statistic: %f' % result[0])
#     print('p-value: %f' % result[1])
    return result[1] <= 0.1  # Check coin is at least 90% stationary
```

# plot of crypto 

```python
# change plot in python file class to get time from question's start to end time
def plot_crypto(tickerDf, tickerSymbol):
    plt.figure(figsize=(10, 4))
    plt.plot(tickerDf.Close)
    plt.title('Stock Price over Time (%s)' % tickerSymbol, fontsize=20)
    plt.ylabel('Price', fontsize=16)
    
    # Set x-axis limits based on start_date and end_date
    plt.xlim(pd.to_datetime(start_date), pd.to_datetime(end_date))
    
    # ploting crypto for specefic time
    for year in range(pd.to_datetime(start_date).year, pd.to_datetime(end_date).year + 1):
        plt.axvline(pd.to_datetime(str(year) + '-01-01'), color='k', linestyle='--', alpha=0.2)

    plt.show()
```

# considering crypto and timeframe

```python
# Fetch and analyze data for each cryptocurrency and timeframe
for crypto in cryptos:
    for timeframe in timeframes:
        # Fetch historical data
        #historical_data = fetch_historical_data(crypto, start_date, end_date, timeframe)
        tickerData = yf.Ticker(crypto)
        tickerDf = tickerData.history(period=timeframe, start=start_date, end=end_date)
        tickerDf = tickerDf[['Close']]

        #ADF test
        is_stationary = perform_adf_test(tickerDf.Close)

        if is_stationary:
            print(tickerDf)
            print(f"{crypto} - {timeframe}: stationary.")
            plot_crypto(tickerDf , crypto)
            

        else:
            print(f"{crypto} - {timeframe}: not stationary.")
```

# part 2

```python
# 'USDT-USD' crypto is stationary according to ADF test
tickerData_stationary = yf.Ticker('USDT-USD')
tickerDf_stationary = tickerData_stationary.history(period=timeframe, start=start_date, end=end_date)
tickerDf_stationary = tickerDf_stationary[['Close']]
```

# acf plot 

```python
acf_plot = plot_acf(tickerDf_stationary.Close)
```

# pacf plot

```python
pacf_plot = plot_pacf(tickerDf_stationary.Close, method='ywm')
```

```python
# 11 monthes for train and 1 month for test
# Set the end dates with timezone information (replace 'UTC' with your actual timezone)
train_end = datetime(2023, 10, 1, tzinfo=timezone.utc)
test_end = datetime(2023, 11, 1, tzinfo=timezone.utc)

# Split the data into training and test sets
train_data = tickerDf_stationary[:train_end]
test_data = tickerDf_stationary.loc[train_end + timedelta(days=1):test_end]
```

# AR model

```python
model = ARIMA(train_data, order=(3,0,0))
```

```python
#fit the model
start = time()
model_fit = model.fit()
end = time()
print('Model fitting time is: ', end-start)
```

```python
#summary of the model
print(model_fit.summary())
```

```python
#get prediction start and end dates
pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]
```

```python
#get the predictions and residuals 
predictions = model_fit.predict(start=pred_start_date, end=pred_end_date)
residuals = test_data.Close - predictions
# print(residuals)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(residuals)
plt.title('Residuals from AR model', fontsize=20)
plt.ylabel('Error', fontsize=16)
plt.axhline(0, color='r', linestyle ='--', alpha=0.2)

# Set x-axis limits based on start_date and end_date
plt.xlim(pd.to_datetime('2023-10-01'), pd.to_datetime('2023-11-02'))

for year in range(pd.to_datetime('2023-10-01').year, pd.to_datetime('2023-11-02').year + 1):
    plt.axvline(pd.to_datetime(str(year) + '-10-01'), color='k', linestyle='--', alpha=0.2)

plt.show()
```

```python
plt.figure(figsize=(10, 4))

# Plot only the data within the specified time period
plt.plot(test_data)
plt.plot(predictions)

plt.legend(('Data', 'Predictions'), fontsize=16)

plt.title('USDT-USD over time (AR)', fontsize=20)
plt.ylabel('Production', fontsize=16)

# Add vertical lines for the years
for year in range(pd.to_datetime('2023-10-01').year, pd.to_datetime('2023-11-02').year + 1):
    plt.axvline(pd.to_datetime(str(year) + '-10-01'), color='k', linestyle='--', alpha=0.2)

plt.show()
```

```python
# Check for zero values in the denominator
non_zero_mask = test_data != 0
print('Mean absolute percent error (MAPE):', round(np.mean(abs(residuals / test_data.Close)),4))
print('Root mean squared error (RMSE):', np.sqrt(np.mean(residuals**2)))
print('mean squared error (MSE):', np.mean(residuals**2))
```

# part 3

```python
import itertools
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# all p and q s
p_values = range(1, 16)
q_values = range(1, 16)

# combining p and q
combinations = list(itertools.product(p_values, q_values))

# we want to save best p and q combination
best_model = None
best_mse = float('inf')
best_mape = float('inf')

for p, q in combinations:
    order = (p, 0, q)  
    
    # training
    model = ARIMA(train_data, order=order)
    model_fit = model.fit()
    
    # prediction
    predictions = model_fit.predict(start=pred_start_date, end=pred_end_date)
    
    # calculation of MAPE and MSE
    residuals = test_data.Close - predictions
    mse = np.mean(residuals**2)
    mape = np.mean(np.abs(residuals / test_data.Close))
    
    # considering the best approach
    if mse < best_mse and mape < best_mape:
        best_model = model_fit
        best_mse = mse
        best_mape = mape
        best_order = order

print('comparision completed')

```

```python
# best approach
print(f"Best Order (p, d, q): {best_order}")
print(f"Best MSE: {best_mse}")
print(f"Best MAPE: {best_mape}")
```

```python
# plot the ACF and PACF
# plot ACF on first plot and PACF on second plot
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(best_model.resid, ax=ax[0])
plot_pacf(best_model.resid, method='ywm', ax=ax[1])
plt.show()
```
